```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
data=pd.read_csv('austin_weather.csv')
data.head()
```

```python
data=data.replace('T',0.0)
data=data.replace('-',0.0)
data['PrecipitationSumInches'] = data['PrecipitationSumInches'].astype(float)
data['RainTomorrow'] = (data['PrecipitationSumInches'] > 0).astype(int)
```

```python
data.head(5)
```

```python
data=data.drop(['Events','Date'],axis=1)
```

```python
for i in data:
    print(data[i].dtype)
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = data[col].astype(float)
```

```python
X = data.drop(
    ['PrecipitationSumInches', 'RainTomorrow'],
    axis=1,
)
y=data['PrecipitationSumInches']  ## this is to predict amt of rain
```

```python
X
```

```python
y_clf = data['RainTomorrow'] ## this to say whether it will rain or not 
```

```python
X_train,X_test,y_train,y_test,y_clf_train, y_clf_test=train_test_split(X,y,y_clf,test_size=0.35,random_state=41)
```

```python
model_pred=LinearRegression()
model_clf=LogisticRegression()
```

```python
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model_pred.fit(X_train,y_train)
model_clf.fit(X_train,y_clf_train)
y_pred = model_pred.predict(X_test)
y_pred_clf=model_clf.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report
)

reg_mse = mean_squared_error(y_test, y_pred)
reg_rmse = np.sqrt(reg_mse)
reg_r2 = r2_score(y_test, y_pred)

print("=== Regression: Predicting Rainfall (inches) ===")
print(f"RMSE: {reg_rmse:.4f} inches")
print(f"R²:   {reg_r2:.4f}\n")

clf_acc = accuracy_score(y_clf_test, y_pred_clf)

print("=== Classification: Predicting Rain / No Rain ===")
print(f"Accuracy: {clf_acc:.4f}")
print("\nClassification report:")
print(classification_report(y_clf_test, y_pred_clf, target_names=["No Rain", "Rain"]))
```
